```python
import os
import csv
import numpy as np
import time

headers=['Name','Path','Identity','Gender_code','Gender','Age','Race_code','Race','date of birth','Emotion_code','Neutral','Anger','Scream','Contempt',
         'Disgust','Fear','Happy','Sadness','Surprise','Sun glasses','Scarf','Eyeglasses','Beard',
         'Hat','Angle']

directory = "root_dir\datasets\original\muct\img"
labels_muct =[]
c=0

for img_name in os.listdir(directory):
    c+=1
    if c==1:
        time_start=time.time()
    
    if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
        img_path = os.path.join(directory, img_name)

        infos = img_name.split('.')[0]
        id = infos[1:4]
        gender = infos[-2]
        glasses = infos[-1]

        label = {
        'Name':img_name,'Path':img_path,'Identity': id,'Gender_code':'','Gender': '', 
        'Age': '','Race_code':'', 'Race': '', 'date of birth': '', 'Emotion_code':'','Neutral': '',
        'Anger': '', 'Scream': '', 'Contempt': '','Disgust': '','Fear': '',
        'Happy': '', 'Sadness': '', 'Surprise': '', 'Sun glasses': '',
        'Scarf': '', 'Eyeglasses': '', 'Beard': '', 'Hat': '', 'Angle': ''}

        if gender == 'm':
            label['Gender_code']=1
            label['Gender']='Male'
        else:
            label['Gender_code']=-1
            label['Gender']='Female'
        if glasses == 'g':
            label['Eyeglasses']=1
        else:
             label['Eyeglasses']=-1
    labels_muct.append(label)
    if c==1:
        end_time = time.time()
        duration_sec = len(os.listdir(directory))*(end_time-time_start)
        duration_hour = duration_sec/3600
        print(f"estimated duration:{duration_hour} hours ~ {duration_sec} sec") 

with open("root_dir\datasets\labels\\muct_labels.csv","w",newline = '') as csv_file:
    csv_writer = csv.DictWriter(csv_file,fieldnames=headers)
    csv_writer.writeheader()
    csv_writer.writerows(labels_muct)

with open("root_dir\datasets\labels\\muct_labels.csv","r",newline = '') as csv_file:
    csv_reader = csv.reader(csv_file)
    c=0
    for line in csv_reader:
        c+=1
        if c%500==0:
            print(line)   
```
